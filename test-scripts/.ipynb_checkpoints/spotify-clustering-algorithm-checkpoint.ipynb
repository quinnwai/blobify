## Wrangling Spotify Data

My first goal is to load in the data in chunks so that I can get all of my liked songs' track IDs. This get request is done using Spotipy, a package that makes Python communication with the Spotify API easier. Stored in my conda environment are the client ID, client secret, and redirect URI that are passed in to the ````spotipy.Spotify```` constructor.

`````python
# GET request for all track ids
scope = "user-library-read"

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(scope=scope))
track_ids = []


SAVED_TRACKS_LIMIT = 20
i = 0

print("num ids:", end=" ")

# for each API request of 20 tracks
while True: 
    response = sp.current_user_saved_tracks(offset=SAVED_TRACKS_LIMIT*i)

    for idx, item in enumerate(response['items']):
        track = item['track']
        track_ids.extend([track['id']])
#         print(idx, track['artists'][0]['name'], " â€“ ", track['name'])

    # progress bar
    if(len(track_ids) % 100 == 0):
        print(len(track_ids), end=" ")
    
    # either end loop once no more tracks to add or increment counter
    if not response['next']:
        break
    
    i += 1
    
print("success!")
    
# ensure that all track ids are unique (not duplicating weirdly)
assert len(track_ids) == len(np.unique(track_ids)), f'Tracks ids not read in properly, as only {len(np.unique(track_ids))}/{len(track_ids)} track IDs are unique'
`````

Then, I need to get associate audio features like acousticness, energy, etc that are associated with each song using the track IDs.

`````python
# get audio features for each track using track_ids

AUDIO_FEATURES_LIMIT = 100
num_requests = int(len(track_ids)/AUDIO_FEATURES_LIMIT)
remainder = len(track_ids)%AUDIO_FEATURES_LIMIT

assert num_requests*AUDIO_FEATURES_LIMIT+remainder == len(track_ids), f'math is wrong: {num_requests*AUDIO_FEATURES_LIMIT+remainder}, {len(track_ids)}'

# initialize lists
audio_features = []

print(f'going to {num_requests}...')

# read in audio features by 100s for all of track_ids (limit is 100)
for i in range(num_requests):
    # create range of track ids to use
    lo = i*AUDIO_FEATURES_LIMIT
    hi = (i+1)*AUDIO_FEATURES_LIMIT
    response = sp.audio_features(tracks=track_ids[lo:hi])

    # add to relevant list
    for j in range(len(response)):
        audio_features.extend([[response[j]['acousticness'], response[j]['danceability'], 
                              response[j]['energy'], response[j]['instrumentalness'], 
                              response[j]['liveness'], response[j]['speechiness'], 
                              response[j]['valence']]])

    # progress bar
    print(i, end =" ")

# get remaining 
if(remainder > 0):
    for j in range(0,remainder):
        audio_features.extend([[response[j]['acousticness'], response[j]['danceability'], 
                              response[j]['energy'], response[j]['instrumentalness'], 
                              response[j]['liveness'], response[j]['speechiness'], 
                              response[j]['valence']]])
    
print("success!")

assert len(track_ids) == len(audio_features), f'length of feature list is {len(audio_features)} while track_ids is {len(track_ids)}'
`````

From there, I take a look at my table and make sure everything seems alright before moving onto qualitative/graphical checks.

`````python
# load all audio features data into dataframe
data = pd.DataFrame(audio_features, columns=['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'speechiness', 'valence'])
data.head(5)
`````

## Determining the Optimal Number of Clusters 

### K Means Algorithm

To determine the optimal cluster of numbers, I chose to use the average silhouette method. The average silhouette method is a helpful metric for my use case, as the optimal number of clusters has the highest average silhouette score. As a result, when I am trying to automate this entire process for users of my website, I can automate the entire algorithmic process by getting the cluster number associated with the maximum score.

`````python
print(__doc__)

avg_silhouette_scores = []

for n_clusters in range(2,21):
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    # The silhouette coefficient can range from -1, 1
    ax1.set_xlim([-1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    # plots of individual clusters, to demarcate them clearly.
    ax1.set_ylim([0, len(data) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # seed of 10 for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(data)

    # Calculate/display silhouette score and add it to list
    avg_silhouette = silhouette_score(data, cluster_labels)
    avg_silhouette_scores.extend([avg_silhouette])
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", avg_silhouette)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(data, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # cluster i, and sort them
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-1,-0.8,-0.6,-0.4,-0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

plt.show()
`````

Through the cluster assignment graph above and the low average silhouette scores, it seems clear that the k-means algorithm here is flawed in some way here. One thing that immediately strikes me here is that the variances are pretty different between each audio feature, as displayed by the previous graphs. Furthermore, we also cannot know that the clusters will have an approximately equal number of datapoints, which might cause the algorithm to select higher. ### DBSCAN Algorithm

Instead of using a centroid-based clustering algorithm, I want to now try a density-based clustering algorithm for comparison. Hopefully, this change will make more useful cluster by removing assumptions about similar variances, number of datapoints per cluster, or spherical cluster shapes. However, one thing to note is that DBSCAN does not work too well for variable density clusters, which will be something to look out for.


Since I'm working with a different set of hyperparameters, I now need to find a way to determine epsilon and minimum number of samples. Using a refererence from Kamil Mysiak's [article](https://towardsdatascience.com/explaining-dbscan-clustering-18eaf5c83b31), I chose to use minimum number of samples as 14, or double my number of features. This will become my nearest neighbors as I analytically try to determine the clusters.

Other source: [Towards Data Science](https://towardsdatascience.com/machine-learning-clustering-dbscan-determine-the-optimal-value-for-epsilon-eps-python-example-3100091cfbc)

`````python
from sklearn.neighbors import NearestNeighbors

# fit model to current data and get all distances from one point to it's closest neighbor
min_samples = data.shape[1]*2
model_neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors = model_neighbors.fit(data)
distances, indices = neighbors.kneighbors(data)

# sort and plot distances
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances, c="darkgreen")
plt.title(f'Sorted Nearest Neighbor Distances')
plt.ylabel("Distances")
plt.xlabel("Datapoints")

# add line to show optimal epsilon
plt.plot(np.repeat(0.2, len(track_ids)), linestyle = '--', c="dimgrey")
plt.show()
`````

Despite what I said earlier about the elbow method for k means, this method using the 'elbow' or highest change in the graph seemed to be the most reliable option. So for now, we will stick with the graphical analysis method to determine the nearest neighbors.

With an epsilon of 0.2, now it's onto fitting the DBSCAN model to predict clusters.

`````python
# fit model with datapoints
dbscan_model = DBSCAN(eps=0.2, min_samples=min_samples)
dbscan_cluster_labels = dbscan_model.fit_predict(data)

np.unique(dbscan_cluster_labels)
`````
